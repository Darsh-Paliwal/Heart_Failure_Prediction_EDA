```python
import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('heart.csv')
```

```python
df.head()
```

```python
df.info()
```

### Data analysis through Distplot

```python
fig , ax = plt.subplots(3,2 , figsize=(12,18))
sns.histplot(df['Age'] , ax=ax[0,0] , kde=True)
sns.histplot(df['RestingBP'] ,ax=ax[0,1], kde=True)
sns.histplot(df['Cholesterol'] ,ax=ax[1,0], kde=True)
sns.histplot(df['FastingBS'] ,ax=ax[1,1], kde=True)
sns.histplot(df['MaxHR'],ax=ax[2,0], kde=True)
sns.histplot(df['Oldpeak'],ax=ax[2,1], kde=True)
plt.show()
```

-> Age graph shows that people between age 40 and 70 are more

-> resting bp shows that value of it mostly lies between 100 - 150

-> In cholestrol people with 0 cholestrol are the most and most of them lie between 150 - 300

-> most of the maxHr value lies between 110 - 170


# Categorical Analysis 

```python
df.groupby(by='Sex').size().plot(kind='pie' , autopct='%.1f')  # 79% is male and 21% is female in this dataset
```

```python
df.groupby(by='ChestPainType').size().plot(kind='pie' , autopct='%.1f')
```

[TA: Typical Angina, ATA: Atypical Angina, NAP: Non-Anginal Pain, ASY: Asymptomatic]

Asymptomatic chest pain has the occuring rate 

```python
df.groupby(by='RestingECG').size().plot(kind='pie' , autopct='%.1f' , title='RestingECG')
```

[Normal: Normal, ST: having ST-T wave abnormality (T wave inversions and/or ST elevation or depression of > 0.05 mV), LVH: showing probable or definite left ventricular hypertrophy by Estes' criteria]

result of ecg and 60% are normal 

```python
df.groupby(by='ExerciseAngina').size().plot(kind='pie' , autopct='%.1f' , title='ExerciseAngina') 
# nearly 40 percent of them having excercise indused angina
```

```python
df.groupby(by='ST_Slope').size().plot(kind='pie' , autopct='%.1f' , title='ST_Slope')
```

the slope of the peak exercise ST segment [Up: upsloping, Flat: flat, Down: downsloping]

50 percent of them having it flat
